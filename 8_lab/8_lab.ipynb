```python
# import sys  
# !{sys.executable} -m pip install --user matplotlib numpy scipy
import csv
import matplotlib.pyplot
import numpy
import scipy.stats
```

```python
data = []
with open('dataset.txt', 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        data.append(row)
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []
for row in data:
    age.append(int(row[0]))
    sex.append(row[1])
    bmi.append(row[2])
    children.append(int(row[3]))
    smoker.append(row[4])
    region.append(row[5])
    charges.append(float(row[6]))
```

```python

fig, ax = matplotlib.pyplot.subplots()

ax.scatter(age, charges)

matplotlib.pyplot.xlabel("Age") 
matplotlib.pyplot.ylabel("Charges") 

matplotlib.pyplot.locator_params(axis='both', nbins=20) 

matplotlib.pyplot.show()
```

```python
smoker_charges = []
smoker_age = []
non_smoker_charges = []
non_smoker_age = []

for person in age:
    if smoker[person] == "yes":
        smoker_charges.append(charges[person])
        smoker_age.append(age[person])
    elif smoker[person] == "no":
        non_smoker_charges.append(charges[person])
        non_smoker_age.append(age[person])

fig, ax = matplotlib.pyplot.subplots()

ax.scatter(smoker_age, smoker_charges, color='red')
ax.scatter(non_smoker_age, non_smoker_charges, color = 'green')

matplotlib.pyplot.xlabel("Age") 
matplotlib.pyplot.ylabel("Charges") 
red_patch = matplotlib.patches.Patch(color='red', label='Smokers')
green_patch = matplotlib.patches.Patch(color='green', label='Non- smokers')
ax.legend(handles=[red_patch, green_patch])

matplotlib.pyplot.locator_params(axis='both', nbins=10) 

matplotlib.pyplot.show()
```

```python
non_child_average_charges = 0
one_child_average_charges = 0
two_child_average_charges = 0
three_child_average_charges = 0
four_child_average_charges = 0
five_child_average_charges = 0
a = 0
b = 0
c = 0
d = 0
e = 0
f = 0
x = [0, 1, 2, 3, 4, 5]

for person in age:
    if children[person]==0:
        non_child_average_charges = non_child_average_charges + charges[person]
        a+=1
    if children[person]==1:
        one_child_average_charges = one_child_average_charges + charges[person]
        b+=1
    if children[person]==2:
        two_child_average_charges = two_child_average_charges + charges[person]
        c+=1
    if children[person]==3:
        three_child_average_charges = three_child_average_charges + charges[person]
        d+=1
    if children[person]==4:
        four_child_average_charges = four_child_average_charges + charges[person]
        e+=1
    else:
        five_child_average_charges = five_child_average_charges  + charges[person]
        f+=1

average_charges = [non_child_average_charges/a, one_child_average_charges/b, two_child_average_charges/c,
                   three_child_average_charges/d, four_child_average_charges/e, five_child_average_charges/f]

fig, ax = matplotlib.pyplot.subplots()

bar_colors = ['crimson', 'orange', 'gold', 'mediumseagreen', 'lightseagreen', 'steelblue']
ax.bar(x, average_charges, width=1, edgecolor="white", linewidth=0.7, color=bar_colors)
matplotlib.pyplot.xlabel("Quantity of children") 
matplotlib.pyplot.ylabel("Charges") 

matplotlib.pyplot.show()
```

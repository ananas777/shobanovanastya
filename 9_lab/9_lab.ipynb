```python
# import sys  
# !{sys.executable} -m pip install --user pandas 
import pandas 
import json
```

```python
df = pandas.read_csv ("ds_salaries.csv")
df.to_json("ds_salaries.json")
```

```python
mean_salaries_by_title = ""
median_salaries_by_title = ""

g1 = df.groupby('job_title', group_keys=False)
for a in g1.groups:
    (g1.get_group(a)).to_json("%s.json" % a)
    mean_salaries_by_title = mean_salaries_by_title + "Mean salary for %s is %f" % (a, g1.get_group(a)["salary_in_usd"].mean()) + "\n"
    median_salaries_by_title = median_salaries_by_title + "Median salary for %s is %f" % (a, g1.get_group(a)["salary_in_usd"].median()) + "\n" 
print(mean_salaries_by_title)
print(median_salaries_by_title)
```

```python
mean_salaries_by_employee_region = ""
median_salaries_by_employee_region = ""

g2 = df.groupby('employee_residence', group_keys=False)
for  a in g2.groups:
    mean_salaries_by_employee_region = mean_salaries_by_employee_region + "Mean salary in %s is %f" % (a, g2.get_group(a)["salary_in_usd"].mean()) + "\n"
    median_salaries_by_employee_region = median_salaries_by_employee_region + "Median salary in %s is %f" % (a, g2.get_group(a)["salary_in_usd"].median()) + "\n"   

print(mean_salaries_by_employee_region)
print(median_salaries_by_employee_region)
    
```

```python
max = 0
result = ""
g3 = df.groupby('job_title', group_keys=False)
for  a in g3.groups:
    if max <= g3.get_group(a)["salary_in_usd"].mean():
        max = g3.get_group(a)["salary_in_usd"].mean()
        result = "Vacancy " + a + " is the highest paying job, because its average salary is: " + str(g3.get_group(a)["salary_in_usd"].mean())
print(result)
```

```python
vacancies = {}
percents = ""
g4 = df.groupby('employee_residence', group_keys=False).cumcount()
for a in g4:
    if df.at[a, 'employee_residence'] in vacancies:
        vacancies[df.at[a, 'employee_residence']] += 1
    else:
        vacancies[df.at[a, 'employee_residence']] = 0
for vacancy in vacancies:
    percents = percents + "Vacancies in %s takes %f%% of whole quantity of vacancies \n" % (vacancy, vacancies[vacancy]*100/len(g4))
print(percents)
```
